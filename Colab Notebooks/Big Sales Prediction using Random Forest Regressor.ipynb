```python
df[['Item_Fat_Content']].value_counts()
```

```python
df.replace({'Item_Fat_Content':{'Low Fat':0,'Regular':1}}, inplace=True)
```

```python
df[['Item_Type']].value_counts()
```

```python
df.replace({'Item_Type':{'Fruits and Vegetables':0,'Snack Foods':0,'Household':1,'Frozen Foods':0,
                         'Dairy':0,'Baking Goods':0,
                         'Canned':0,'Health and Hygiene':1,
                         'Meat':0,'Soft Drinks':0, 'Breads':0,
                         'Hard Drinks':0, 'Others':2, 'Starchy Foods':0,
                         'Breakfast':0, 'Seafood':0}}, inplace=True)
```

```python
df[['Item_Type']].value_counts()
```

```python
df[['Outlet_Identifier']].value_counts()
```

```python
df.replace({'Outlet_Identifier':{'OUT027':0,'OUT013':1,
                                 'OUT049':2,'OUT046':3,'OUT035':4,
                                 'OUT045':5,'OUT018':6,
                                 'OUT017':7,'OUT010':8,'OUT019':9}},inplace=True)
```

```python
df[['Outlet_Identifier']].value_counts()
```

```python
df[['Outlet_Size']].value_counts()
```

```python
df.replace({'Outlet_Size':{'Small':0,'Medium':1,'High':2}},inplace=True)
```

```python
df[['Outlet_Size']].value_counts()
```

```python
df[['Outlet_Location_Type']].value_counts()
```

```python
df.replace({'Outlet_Location_Type': {'Tier 1':0,'Tier 2':1,'Tier 3':2}}, inplace=True)
```

```python
df[['Outlet_Location_Type']].value_counts()
```

```python
df[['Outlet_Type']].value_counts()
```

```python
df.replace({'Outlet_Type': {'Grocery Store':0, 'Supermarket Type1':1,
                            'Supermarket Type2':2,'Supermarket Type3':3,
                            }}, inplace=True)
```

```python
df[['Outlet_Type']].value_counts()
```

```python
df.info()
```

```python
df.shape
```

```python
y=df['Item_Outlet_Sales']
```

```python
y.shape
```

```python
y
```

```python
df.columns
```

```python
X=df[['Item_Weight', 'Item_Fat_Content', 'Item_Visibility',
       'Item_Type', 'Item_MRP', 'Outlet_Identifier',
       'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type',
       'Outlet_Type']]
```

```python
X.shape
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_std= df[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
```

```python
X_std =sc.fit_transform(X_std)
```

```python
X_std
```

```python
X[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]=pd.DataFrame(X_std, columns=[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.1,random_state=2529)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr= RandomForestRegressor(random_state=2529)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred= rfr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
```
