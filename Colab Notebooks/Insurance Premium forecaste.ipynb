```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/Insurance%20Premium.csv')
```

```python
df.head()
```

```python
df.replace({'Gender':{'male': 0, 'female':1}},inplace=True)
#inplace=True dataframe permanent change
```

```python
df.replace({'Smoker':{'no':0,'yes':1}},inplace=True)
```

```python
df.replace({'Region':{'east':0,'west':1,'north':2,'south':3}},inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
X=df[['Age', 'Gender', 'BMI', 'Children', 'Smoker', 'Region',
       ]]
```

```python
# since gender, smoker is in string- replace in 0 & 1
# scaling is also done for other than categorical var
```

```python
X.shape
```

```python

y=df['Premium']
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler() 
```

```python
X_std=df[['Age','BMI']]
```

```python
X_std = sc.fit_transform(X_std)
```

```python
X_std
```

```python
X[['Age','BMI']]=pd.DataFrame(X_std,columns=['Age','BMI'])
```

```python
X
```

```python
# use dummy variable for Region to see impact analysis. 
# if purpose is forecast then generally not big impact
# if policy making then impact 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=252,)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(random_state=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
# explaination power of model
```
