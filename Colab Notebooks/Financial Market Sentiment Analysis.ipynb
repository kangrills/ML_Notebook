```python
# in text analytics there is gram
# Today is a cold day (statement)
# word, token and gram  are synonyms. we need bag of words
# analysis can be done n-gram ie unigram, bigram etc
# bigram approach: Today is, is a , cold day
# Trigram approach: Today is a, is a cold, a cold day etc
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# countVectorizer: bag of words frequency create
```

```python
cv = CountVectorizer(lowercase = True, ngram_range=(1,1))
```

```python
# bag of word approach we see frequency of words. - countvectorizer
# assign sentiment +1,-1 or 0 to the words
# all words have degree of positive or degree of negative
```

```python
X = cv.fit_transform(X)
```

```python
X.shape
# 4101- rows
# 48527- column. Tells no of words (bag of words)
```

```python
y= df["Label"]
```

```python
y.shape
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=252,)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(random_state=252)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
