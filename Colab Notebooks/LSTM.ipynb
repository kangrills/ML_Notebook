```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv(r'/content/NHPC.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
training_set=df[['Energy']] #for min max scalar
training_set.shape
#double bracket to make it column vector
# y is column vector and X(features) as row vector
```

```python
# for every ANN problem we will do standardization if not done the model is misleading
```

```python
# we will use min max scalar as nothing to do with distribution. In time series always use min max scalar
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler(feature_range=(0,1))
#assign values between 0 and 1
```

```python
training_set_scaled=sc.fit_transform(training_set)
```

```python
X_train = []  #create empty list
y_train = [] 
for i in range(60, 1258): 
  X_train.append(training_set_scaled[i-60:i, 0])
  #0 is column name
  y_train.append(training_set_scaled[i, 0]) 
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
#recurssive window is 60
```

```python
y_train
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
from keras.models import Sequential #line by line
```

```python
from keras.layers import Dense #all are fully conected
```

```python
from keras.layers import LSTM
```

```python
from keras.layers import Dropout 
#Regularization of neurons to prevent overfiting ie if a neuron is dead, make neurons general specific. avoid specialization
```

```python
regressor=Sequential()
```

```python
regressor.add(LSTM(units=50,return_sequences=True, input_shape=(X_train.shape[1],1)))
#units means  no of neurons, sequential data (time series) not on standalone data
```

```python
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))
```

```python
regressor.add(Dense(units=1))
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressor.fit(X_train, y_train, epochs = 20, batch_size = 32)
```

```python
X_train.shape
```

```python
predicted=regressor.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error
```
