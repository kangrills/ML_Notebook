#Q1
 A **simple random sample** is  a purest form of probability sampling and used to represent the entire data population and randomly selects individuals from the population without any other consideration.

In simple random sampling each member has the equal chance of selection.

Example: A company wants to study the overall user experience of a product by its customers then it will randomly select 200 customers and record their feedback.

A **stratified random sampling** is obtained by dividing the population elements into non overlapping groups, called strata and selecting a random sample directly and independently from each stratum.

Stratas are a subset of a population that shares atleast one common characteristic.

Example: A candy company may decide to use the random stratified sampling method by dividing its 100 customers into different age groups to help make determinations about the future of its production.

A Research method can be selected based on the type of the problem. If the problem is forecasting / prediction / estimation then we will use machine learning.
It can be of 2 types:

1.supervised Machine learning.

2.unsupervised Machine learning.

**Supervised Machine learning** is used when the data of independent variable is given.

Example: Predicting life expactancy based on age, gender, alcohol comsumption.

**Unsupervised Machine learning** is used when data of independent variable is not given.

Example: image classification problem in the computer vision field predicting what class an image belongs to.


**Supervised Machine learning**
1. In case of Supervised Machine learning if the independent variable is continous we will use regression methods like simple regression etc , 
example: forecast stock returns based on index

2. In Supervised Machine learning if the independent variable is categorical we will use Classification methods like logistic regression etc , example: Default or no default based on cibil score of client.

```python
lr=LogisticRegression()
```
